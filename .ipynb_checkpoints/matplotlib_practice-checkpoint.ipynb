```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the X-axis to a list of strings for each month
x_axis=["Jan","Feb","Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]

#Set the Y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the same plot using the object-oriented interface with ax.plt()
fig, ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create thw plot a different way
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Another way
ax=plt.axes()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and add a label for the ledgend
plt.plot(x_axis,y_axis, label="Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the ledgend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis,y_axis,marker="*",color="blue",linewidth=2,label="Boston")

#Create the labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()

# Add the ledgend
plt.legend()
```

```python
# Creating a bar chart. Set the x-axis to a list of strings for each month
x_axis=["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis,y_axis)
```

```python
# Create the plot
plt.bar(x_axis,y_axis, label="Boston")

# Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis,y_axis)
```

```python
# Inver the order of the months
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
# Create bar chart using the object-oriented method
# Set the x_axis to a list of strings for each month
x_axis=["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]

# Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plot()
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis)
```

```python
# Create horizontal bar chart with object-oriented interface
# Create the the plot with ax.plt()
fig,ax=plt.subplots()
ax.barh(x_axis,y_axis, color="cyan")
```

```python
# Create a scatterplot using the MATLAB method
# Set the x-axis to a list of strings for each month
x_axis=["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]

# Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the scatter plot
plt.plot(x_axis,y_axis,"o")
```

```python
plt.scatter(x_axis,y_axis, c="red",)
```

```python
# Create bubble chart using MATLAB
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
# Make the points larger
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
# Using list comprehension technique to mult y_axis data points
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Create scatter plot using object-oriented interface
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Create the bubble chart
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis,s=y_axis)
```

```python
# Create a MATLAB pie chart
plt.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Increase the size of the pie chart
plt.subplots(figsize=(8,8))
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode=explode_values,labels=x_axis, autopct="%.1f%%")
```

```python
# Changing the colors of the pieces of the pie
#Assign 12 colors, one for each month
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow","yellow","orange","gold","indianred","tomato","mistyrose"]
explode_values=[0,0,0,0,0,0,0.2,0,0,0,0,0]
plt.pie(y_axis,explode=explode_values,colors=colors,labels=x_axis,autopct="%.1f%%")
plt.show()
```
